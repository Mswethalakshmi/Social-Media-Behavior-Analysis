```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data
df = pd.read_csv('Dataset  (1).csv')
```

```python
# Display the first few rows
print(df.head())         
```

```python
# Get info about the DataFrame
print(df.info())  
```

```python
# Get statistical summary
print(df.describe())     
```

```python
# Check for missing values
df.isnull().sum()       
```

```python
# Drop rows with missing values
df.dropna()
```

```python
# Split and explode cuisines column
df_cuisines = df.assign(Cuisines=df['Cuisines'].str.split(', ')).explode('Cuisines')

# Count and get top 3 cuisines
top_cuisines = df_cuisines['Cuisines'].value_counts().head(3)

# Calculate percentages
top_cuisines_percentage = (top_cuisines / len(df)) * 100
print("Top 3 cuisines:\n", top_cuisines)
print("\nPercentage of restaurants serving top cuisines:\n", top_cuisines_percentage)
```

```python
city_counts = df['City'].value_counts()
top_city = city_counts.idxmax()
print("City with highest number of restaurants:", top_city)
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean()
print("Average rating for each city:\n", city_avg_rating)
```

```python
top_rated_city = city_avg_rating.idxmax()
print("City with highest average rating:", top_rated_city)
```

```python
# Bar chart for price range distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Price range', data=df)
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
price_range_counts = df['Price range'].value_counts()
price_range_percentage = (price_range_counts / len(df)) * 100
print("Percentage of restaurants in each price range:\n", price_range_percentage)
```

```python
online_delivery_percentage = (df['Has Online delivery'].value_counts(normalize=True) * 100)
print("Percentage of restaurants offering online delivery:\n", online_delivery_percentage)
```

```python
avg_rating_with_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
avg_rating_without_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()
print("Average rating with online delivery:", avg_rating_with_delivery)
print("Average rating without online delivery:", avg_rating_without_delivery)
```

```python
# Histogram for aggregate ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=10)
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
most_common_rating = df['Aggregate rating'].mode()[0]
avg_votes = df['Votes'].mean()
print("Most common rating range:", most_common_rating)
print("Average number of votes:", avg_votes)
```

```python
chains = df['Restaurant Name'].value_counts()
popular_chains = chains[chains > 1]  # Identify chains with more than one location
print("Restaurant chains:\n", popular_chains)

# Analyze ratings of chains
chain_ratings = df[df['Restaurant Name'].isin(popular_chains.index)].groupby('Restaurant Name')['Aggregate rating'].mean()
print("Ratings for restaurant chains:\n", chain_ratings)
```

```python
top_votes = df.nlargest(10, 'Votes')
bottom_votes = df.nsmallest(10, 'Votes')
print("Restaurants with highest votes:\n", top_votes)
print("Restaurants with lowest votes:\n", bottom_votes)

# Correlation between votes and ratings
correlation = df[['Votes', 'Aggregate rating']].corr()
print("Correlation between votes and rating:\n", correlation)
top_votes = df.nlargest(10, 'Votes')
bottom_votes = df.nsmallest(10, 'Votes')
print("Restaurants with highest votes:\n", top_votes)
print("Restaurants with lowest votes:\n", bottom_votes)

# Correlation between votes and ratings
correlation = df[['Votes', 'Aggregate rating']].corr()
print("Correlation between votes and rating:\n", correlation)
```

```python
# Relationship between price range and online delivery
online_delivery_price = pd.crosstab(df['Price range'], df['Has Online delivery'])
table_booking_price = pd.crosstab(df['Price range'], df['Has Table booking'])
print("Online delivery by price range:\n", online_delivery_price)
print("Table booking by price range:\n", table_booking_price)

# Higher-priced restaurants
high_price_online = df[df['Price range'] == 4]['Has Online delivery'].value_counts(normalize=True) * 100
print("Percentage of high-priced restaurants offering online delivery:\n", high_price_online)
```
