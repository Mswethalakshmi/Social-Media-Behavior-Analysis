```python
data['how many hours per day do you spend online?'] = pd.to_numeric(data['how many hours per day do you spend online?'], errors='coerce')
data['what is the total number of social media platforms that you use?'] = pd.to_numeric(data['what is the total number of social media platforms that you use?'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar chart for "what is your occupation?"
plt.figure(figsize=(12, 6))
sns.countplot(x='what is your occupation?', data=data, palette='viridis')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Distribution of Occupations')
plt.xticks(rotation=45)
plt.show()
```

```python
# Horizontal bar chart for "what device do you use most to access the internet?"
plt.figure(figsize=(10, 5))
sns.countplot(y='what device do you use most to access the internet?', data=data, palette='viridis')
plt.ylabel('Primary Device Used')
plt.xlabel('Count')
plt.title('Distribution of Primary Devices Used')
plt.show()
```

```python
device_counts = data['what device do you use most to access the internet?'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(device_counts)))
plt.title('Primary Devices Used')
plt.show()
```

```python
# Pair plot for numerical columns
sns.pairplot(data)
plt.title('Pairwise Relationships')
plt.show()
```

```python
premium_subscription_counts = data['do you purchase premium subscriptions for social media platforms? (like spotify or netflix premium)'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(premium_subscription_counts, labels=premium_subscription_counts.index, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(premium_subscription_counts)))
plt.title('Purchase of Premium Subscriptions')
plt.show()
```

```python
from textblob import TextBlob

# Ensure the correct column name is used (in quotes)
opinion_column = 'do you feel like attention is being monetized by platforms?'  # Replace with the correct column name

# Example: Sentiment analysis on the text column
data['sentiment'] = data[opinion_column].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Bar chart for sentiment analysis
plt.figure(figsize=(12, 6))
sns.histplot(data['sentiment'], bins=20, kde=True, color='skyblue')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.title('Distribution of Sentiment Polarity')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='category', data=data, palette='viridis')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.xticks(rotation=45)
plt.show()
```
